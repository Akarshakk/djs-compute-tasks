```python
# UFC Fighters Statistics - Data Cleaning & EDA

# =============================
# Import Required Libraries
# =============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")

# =============================
# Load Dataset
# =============================
df = pd.read_csv("ufc-fighters-statistics.csv") #reads the csv into a dataframe df 

print("Initial Shape:", df.shape) #number of rows and columns
print("\nFirst 5 rows:")
display(df.head()) #first 5 rws review

# =============================
# Data Cleaning
# =============================

# 1. Remove duplicates
df = df.drop_duplicates() #checks all the contents of the column
df = df.drop_duplicates(subset=["name"], keep="first") #just checks the "name" column

# 2. Handle missing values
for col in df.select_dtypes(include=[np.number]).columns:
    df[col] = df[col].fillna(df[col].median()) #replaces all missing values in that numeric attribute column with the median

for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].fillna("Unknown") #replaces all missing values in that string attribute column with "unknown"

# 3. Convert date_of_birth to datetime & create age feature
if "date_of_birth" in df.columns:
    df["date_of_birth"] = pd.to_datetime(df["date_of_birth"], errors="coerce") #converts DOB to date-time format
    df["age_in_years"] = (pd.to_datetime("today") - df["date_of_birth"]).dt.days // 365 #age in years calculation
    df["age_in_years"] = df["age_in_years"].fillna(df["age_in_years"].median()) #replaces the missing ages with median

# 4. Outlier handling (winsorize numeric distributions)
def winsorize(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return np.clip(series, lower, upper) #replaces the outliers with lower and upper values

for col in ["height_cm", "weight_in_kg", "reach_cm"]:
    if col in df.columns:
        df[col] = winsorize(df[col])

# 5. Clip percentage metrics to valid ranges [0,100]
for col in df.columns:
    if "accuracy" in col or "defense" in col:
        df[col] = df[col].clip(0, 100) #shrinks the values in a limit of 0-100

# =============================
# Feature Engineering
# =============================
if {"wins", "losses", "draws"}.issubset(df.columns):
    df["total_fights"] = df["wins"] + df["losses"] + df["draws"] #calculates total fights
    df["win_rate"] = (df["wins"] / df["total_fights"] * 100).replace([np.inf, np.nan], 0) #replaces infinite and null values with 0

# =============================
# Exploratory Data Analysis
# =============================

# --- Height distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df["height_cm"], bins=30, kde=True)
plt.title("Height Distribution of Fighters")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
plt.show()

# --- Weight distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df["weight_in_kg"], bins=30, kde=True, color="orange")
plt.title("Weight Distribution of Fighters")
plt.xlabel("Weight (kg)")
plt.ylabel("Count")
plt.show()

# --- Top 10 fighting stances ---
if "stance" in df.columns:
    plt.figure(figsize=(8,5))
    df["stance"].value_counts().head(10).plot(kind="bar", color="teal")
    plt.title("Top 10 Stances of Fighters")
    plt.xlabel("Stance")
    plt.ylabel("Number of Fighters")
    plt.show()

# --- Height vs Reach ---
if {"height_cm", "reach_in_cm"}.issubset(df.columns):
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="height_cm", y="reach_in_cm", data=df, alpha=0.59)
    plt.title("Height vs Reach")
    plt.xlabel("Height (cm)")
    plt.ylabel("Reach (cm)")
    plt.show()

# --- Total fights vs Win rate ---
if {"total_fights", "win_rate"}.issubset(df.columns):
    plt.figure(figsize=(8,5))
    sns.scatterplot(x="total_fights", y="win_rate", data=df, alpha=0.6, color="purple")
    plt.title("Total Fights vs Win Rate")
    plt.xlabel("Total Fights")
    plt.ylabel("Win Rate (%)")
    plt.show()

# --- Correlation heatmap ---
plt.figure(figsize=(10,6))
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# =============================
# Save Cleaned Data
# =============================
df.to_csv("ufc_fighters_cleaned.csv", index=False)
print("Cleaned dataset saved as 'ufc_fighters_cleaned.csv'")
```
